```python
#import Library
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from pycaret.regression import *
```

```python
#load the Data
from google.colab import files
uploaded = files.upload()
```

```python
# Store the data into dataframes
df_btc = pd.read_csv('BTC-USD.csv')
df_btc
```

```python
#variabel untuk memprediksi 1 hari ke depan
future_days =1
#membuat kolom baru (target atau variabel dependen) menggeser unit 'n' ke atas, dalam hal ini n = 1
df_btc['Future_Price'] = df_btc[['Close']].shift(-future_days)
#pilih kolom yang akan disimpan
df_btc = df_btc[['Close','Future_Price']]
#show the data
df_btc
```

```python
#membuat salinan data
df = df_btc.copy()
#membuat kumpulan data independen
X = np.array(df[df.columns])
#menghapus baris 'n' terakhir dari kumpulan data di mana "n" adalah future_days=1
X= X[:len(df_btc)-future_days]
#membuat kumpulan data dependen (y) dan ubah kerangka data menjadi array numpy
y = np.array(df['Future_Price'])
#memdapatkan semua nilai y kecuali untuk baris 'n' terakhir yang merupakan future_days = 1
y = y[:-future_days]
#pisahkan data menjadi kumpulan data 85% traing dan kumpulan data 15% testing
x_train, x_test, y_train, y_test = train_test_split(X,y, test_size=0.15, random_state =0, shuffle=False)
```

```python
#get the train data and transform it to a dataframe
train_data = pd.DataFrame(x_train, columns= df.columns)
#show the first 7 rows of data
train_data.head(7)
```

```python
#get the train data and transform it to a dataframe
test_data = pd.DataFrame(x_test, columns= df.columns)
#show the first 7 rows of data
test_data.head(7)
```

```python
#inisialisasi pengaturan
regression_setup = setup(data=train_data, target = 'Future_Price', session_id=123,use_gpu = True)
```

```python
#train on all the models and sort it by the R-squared metric aka (r2) and then store the model with the best r2 score
best_model = compare_models(sort='r2')
```

```python
#create the model and show its prediction metrics on the trainging data set
model = create_model(best_model)
```

```python
#Evaluate the model
evaluate_model(model)
```

```python
#Get the predictions
unseen_predictions = predict_model(model, data=test_data)
#show the prediction
unseen_predictions
```
